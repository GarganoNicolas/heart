# 2. Ver columnas que son object

```python
data4=data5.copy()
```

### 2.1 Relativizar el racismo

```python
data4['Race'] = data4['Race'].replace({'White': 'Gringo', 'Black': 'Afro'})
data4['Race'].unique()
```

### 2.2 Pasar de float a int

```python
data4[['SleepTime', 'MentalHealth', 'PhysicalHealth']] = data4[['SleepTime', 'MentalHealth', 'PhysicalHealth']].astype(int)
```

```python
status(data4)
```

# 3. OHE & Yes No columns

```python
data3=data4.copy()
```

```python
for columna in data4:
    # Verificar si el tipo de dato es "object"
    if data4[columna].dtype == 'object':
        # Obtener los valores nicos de la columna
        valores_unicos = data4[columna].unique()

        # Imprimir los valores nicos de la columna con nmeros secuenciales
        print(f"Valores nicos de la columna {columna}:")
        if len(valores_unicos) < 6:
            for indice, valor in enumerate(valores_unicos):
                print(f"{indice}. {valor}")
        print()
```

```python
data4_reemplazado = data4.copy()

# Iterar sobre las columnas del DataFrame
for columna in data4_reemplazado:
    # Verificar si el tipo de dato es "object"
    if data4_reemplazado[columna].dtype == 'object':
        # Obtener los valores nicos de la columna
        valores_unicos = data4_reemplazado[columna].unique()
        if len(valores_unicos) < 6:
            # Crear un diccionario de reemplazo
            diccionario_reemplazo = {valor: indice for indice, valor in enumerate(valores_unicos)}

            # Reemplazar los valores en la columna
            data4_reemplazado[columna] = data4_reemplazado[columna].map(diccionario_reemplazo)

# Imprimir el DataFrame con los valores reemplazados
print(data4_reemplazado)
```

```python
for columna in data4_reemplazado.columns:
    valores_unicos = data4_reemplazado[columna].unique()
    if len(valores_unicos) < 7:
        print(f"Valores nicos en la columna {columna}: {valores_unicos}")
```

```python
data2 = data4_reemplazado
status(data2)
```

```python
data_ohe = pd.get_dummies(data2, dtype=int)
status(data_ohe)
```


# 3. Guardamos el dataFrame

```python
filename = "data_ohe.csv"
data_ohe.to_csv(filename, index = False)
```

```python
# Labels

import numpy as np
columnas4 = data4.columns
for columna in columnas4:


    valor_mas_frecuente = data4[columna].value_counts().idxmax()
    # Verifica el tipo de dato de la columna
    if np.issubdtype(data4[columna].dtype, np.number):
        if len(data4[columna].unique()) <= 5:
            codigo_columna = f"{columna} = gr.Radio(\n"
            codigo_columna += f"    label='{columna.title()}',\n"
            codigo_columna += f"    choices={data4[columna].unique().tolist()},\n"
            codigo_columna += f"    value={valor_mas_frecuente},\n"
            codigo_columna += f")\n"
            print(codigo_columna)
        else:
            codigo_columna = f"{columna} = gr.Slider(\n"
            codigo_columna += f"    label='{columna.title()}',\n"
            codigo_columna += f"    minimum={data4[columna].min()},\n"
            codigo_columna += f"    maximum={data4[columna].max()},\n"
            codigo_columna += f"    step=1,\n"
            codigo_columna += f"    randomize=True\n"
            codigo_columna += f")\n"
            print(codigo_columna)
    else:
        # Verifica la cantidad de valores nicos en la columna
        if len(data4[columna].unique()) <= 3:
            codigo_columna = f"{columna} = gr.Radio(\n"
            codigo_columna += f"    label='{columna.title()}',\n"
            codigo_columna += f"    choices={data4[columna].unique().tolist()},\n"
            codigo_columna += f"    value='{valor_mas_frecuente}',\n"
            codigo_columna += f")\n"
            print(codigo_columna)
        else:
            codigo_columna = f"{columna} = gr.Dropdown(\n"
            codigo_columna += f"    label='{columna.title()}',\n"
            codigo_columna += f"    choices={data4[columna].unique().tolist()},\n"
            codigo_columna += f"    multiselect=False,\n"
            codigo_columna += f"    value='{valor_mas_frecuente}',\n"
            codigo_columna += f")\n"
            print(codigo_columna)
```
